# Lineal Systems pt. 1

## First Execersice
$$y^{'}= Ay$$
$$A = \begin{bmatrix}
-1 & 1\\
-5 & -5
\end{bmatrix} $$

$\textit{Initial Value Problem}$ 
$$y(0) = \begin{bmatrix}
1\\
5
\end{bmatrix} $$

```python
from DE import lineal_system
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
IVP = [1, -5]
```

```python
# Time line
t = np.linspace(0, 30, 60) 
```

```python
# solve ODEs
x1, y1, x2, y2 = -1, 1, -5, -5 
y = odeint(lineal_system,IVP,t, args=(x1, y1, x2, y2,))
```

```python
# Plot x-axis
x_axis = y[:,0]
plt.semilogy(t, x_axis)
plt.xlabel('time')
plt.ylabel('x_axis')
plt.legend()
plt.show()
```

```python
# Plot y-axis
y_axis = y[:,1]
plt.semilogy(t, y_axis)
plt.xlabel('time')
plt.ylabel('y_axis')
plt.legend()
plt.show()
```

```python
# Plot both
# plot results
plt.plot(t,x_axis,'r-',linewidth=2,label='x-axis')
plt.plot(t,y_axis,'b-',linewidth=2,label='y_axis')
plt.xlabel('time')
plt.show()
```

```python
# Slope fields
# Solution curve

# Vector field
X, Y = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))
U = x1*X + y1*Y
V = x2*X + y2*Y
# Normalize arrows
N = np.sqrt(U ** 2 + V ** 2)
U = U / N
V = V / N
plt.quiver(X, Y, U, V, angles="xy")
plt.plot(y[:, 0], y[:, 1], "-")

plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

## Second Execersice
$$y^{'}= Ay$$
$$A = \begin{bmatrix}
-5 & 1\\
-2 & -2
\end{bmatrix} $$

$\textit{Initial Value Problem}$ 
$$y(0) = \begin{bmatrix}
0\\
-1
\end{bmatrix} $$

```python
IVP = [0, -1]
```

```python
# Time line
t = np.linspace(0, 30, 60)
```

```python
# solve system
x1, y1, x2, y2 = -5, 1, -2, -2 
y = odeint(lineal_system,IVP,t, args=(x1, y1, x2, y2,))
```

```python
# Plot x-axis
x_axis = y[:,0]
plt.semilogy(t, x_axis)
plt.xlabel('time')
plt.ylabel('x_axis')
plt.legend()
plt.show()
```

```python
# Plot y-axis
y_axis = y[:,1]
plt.semilogy(t, y_axis)
plt.xlabel('time')
plt.ylabel('y_axis')
plt.legend()
plt.show()
```

```python
# Plot both
# plot results
plt.plot(t,x_axis,'r-',linewidth=2,label='x-axis')
plt.plot(t,y_axis,'b-',linewidth=2,label='y_axis')
plt.xlabel('time')
plt.show()
```

```python
# Slope fields
# Solution curve

# Vector field
X, Y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
U = x1*X + y1*Y
V = x2*X + y2*Y
# Normalize arrows
N = np.sqrt(U ** 2 + V ** 2)
U = U / N
V = V / N
plt.quiver(X, Y, U, V, angles="xy")
plt.plot(y[:, 0], y[:, 1], "-")

plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

## See multiple curve solutions

```python

# Slope fields
# Solution curve

# Vector field
X, Y = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))
U = x1*X + y1*Y
V = x2*X + y2*Y
# Normalize arrows
N = np.sqrt(U ** 2 + V ** 2)
U = U / N
V = V / N
plt.quiver(X, Y, U, V, angles="xy")

for y0 in np.linspace(-5.0, 0.0, 10):
    y_initial = [y0, -10.0]
    y = odeint(lineal_system,y_initial,t, args=(x1, y1, x2, y2,))
    plt.plot(y[:, 0], y[:, 1], "-")
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```
