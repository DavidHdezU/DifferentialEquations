```python
from DE import predator_prey_model
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initial condition
x0 = [4, 4]
```

```python
# Time line
t = np.linspace(0, 30, 60) 
```

```python
# solve ODEs
a, b, c, d = 0.16, 0.08, 0.9, 4.5 
y1 = odeint(predator_prey_model,x0,t, args=(a,b,c,d,))
```

```python
# Plot predator
predators = y1[:,0]
plt.semilogy(t, predators)
plt.xlabel('time')
plt.ylabel('Predators')
plt.legend()
plt.show()
```

```python
# Plot preys
preys = y1[:,1]
plt.semilogy(t, preys)
plt.xlabel('time')
plt.ylabel('Preys')
plt.legend()
plt.show()
```

```python
# Plot both
# plot results
plt.plot(t,predators,'r-',linewidth=2,label='Predators')
plt.plot(t,preys,'b-',linewidth=2,label='Preys')
plt.xlabel('time')
plt.show()
```

```python
# Slope fields
# Solution curve

# Vector field
X, Y = np.meshgrid(np.linspace(0, 20, 20), np.linspace(0, 20, 20))
U = X*(-a + b*Y)
V = Y*(d - c*X)
# Normalize arrows
N = np.sqrt(U ** 2 + V ** 2)
U = U / N
V = V / N
plt.quiver(X, Y, U, V, angles="xy")



plt.plot(y1[:, 0], y1[:, 1], "-")

plt.xlim([0, 20])
plt.ylim([0, 20])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```
